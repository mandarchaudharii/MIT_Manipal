# Manual Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data.xlsx")

df
```

```python
x=np.array(df['Weeks'])
y=np.array(df['Sales'])
```

```python
x
```

```python
y
```

```python
plt.title("Diagram")
plt.scatter(x, y, marker='o')

for i, j in zip(x, y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')

plt.show()
```

```python
x2 = x*x

xy = x*y
```

```python
sum_x = x.sum()

sum_y = y.sum()

sum_x2 = x2.sum()

sum_xy = xy.sum()

print('sum_x = ', sum_x)
print('sum_y = ', sum_y)
print('sum_x2 = ', sum_x2)
print('sum_xy = ', sum_xy)
```

```python
x_bar = sum_x/len(x)

y_bar = sum_y/len(y)

x2_bar = sum_x2/len(x2)

xy_bar = sum_xy/len(xy)

print('x_bar = ', x_bar)
print('y_bar = ', y_bar)
print('x2_bar = ', x2_bar)
print('xy_bar = ', xy_bar)
```

```python
a1 = (xy_bar-(x_bar*y_bar))/(x2_bar-(x_bar*x_bar))

a0 = y_bar - (a1*x_bar)

round(a1, 2)
round(a0, 2)

print('a1 = ' , a1)
print('a0 = ' , a0)
```

```python
#Regression Equation is

print('y =', a0 ,'+', a1,'* x')
```

```python
plt.title("With Regression Equation")
          
y_pred = a0 + (a1*x)

plt.scatter(x,y)
plt.plot(x,y_pred , color ="red")

for i, j in zip(x, y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

```python
#Prediction

print('Prediction For x=7' , a0 + (a1*7))
print('Prediction For x=12' , a0 + (a1*12))
```
