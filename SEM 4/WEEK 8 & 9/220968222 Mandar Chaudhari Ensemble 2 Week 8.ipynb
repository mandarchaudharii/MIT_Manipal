```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("breast_cancer.csv")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
string =['Id', 'Clump_thickness', 'Uniformity_Cell_Size','Uniformity_Cell_Shape', 'Marginal_Adhesion','Single_Epithelial_Cell_Size', 'Bare_Nuclei', 'Bland_Chromatin','Normal_Nucleoli', 'Mitoses', 'Class']
j=0
for i in df:
    df.rename(columns={i: string[j]},inplace=True)
    j+=1
df
```

```python
df = df.drop('Id', axis=1)
```

```python
df['Bare_Nuclei'] =pd.to_numeric(df['Bare_Nuclei'], errors='coerce').convert_dtypes() 
```

```python
df.isnull().sum()
df=df.dropna()
```

```python
Y = df['Class']


X = df.drop(['Class'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
cols = x_train.columns
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

# MAX VOTING

```python
from sklearn.ensemble import VotingClassifier
model1 = KNeighborsClassifier(n_neighbors=5)
model2 = DecisionTreeClassifier(random_state=1)
model = VotingClassifier(estimators=[('lr', model1), ('dt', model2)], voting='hard')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
y_pred_test= model.predict(x_test)
y_pred_train= model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_test)

print(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_test))
```

# AVERAGING

```python
model1 = KNeighborsClassifier(n_neighbors=5)
model2 = DecisionTreeClassifier(random_state=1)

model1.fit(x_train,y_train)
model2.fit(x_train,y_train)

pred1=model1.predict_proba(x_test)
pred2=model2.predict_proba(x_test)

finalpred=(pred1+pred2)/2

finalpred
```

```python
x_test.shape
```

```python
y_pred_test=[]
for i in finalpred:
    if(i[0]>=0.5):
        y_pred_test.append(2)
    elif(i[1]>0.5):
        y_pred_test.append(4)
```

```python
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

```python
cm = confusion_matrix(y_test, y_pred_test)

print(cm)
```

```python
print(classification_report(y_test, y_pred_test))
```

# WEIGHTED AVERAGE

```python
model1 = KNeighborsClassifier(n_neighbors=5)
model2 = DecisionTreeClassifier(random_state=1)

model1.fit(x_train,y_train)
model2.fit(x_train,y_train)

pred1=model1.predict_proba(x_test)
pred2=model2.predict_proba(x_test)

finalpred=(pred1*0.4+pred2*0.6)

finalpred
```

```python
y_pred_test=[]
for i in finalpred:
    if(i[0]>=0.5):
        y_pred_test.append(2)
    elif(i[1]>0.5):
        y_pred_test.append(4)
```

```python
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

```python
cm = confusion_matrix(y_test, y_pred_test)

print(cm)
```

```python
print(classification_report(y_test, y_pred_test))
```
