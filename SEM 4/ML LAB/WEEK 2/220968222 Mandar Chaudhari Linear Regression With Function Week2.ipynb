# Linear Regression With Functions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("data.xlsx")

df
```

```python
reg=LinearRegression()
```

```python
x=np.array(df[['Weeks']])
y=np.array(df[['Sales']])
```

```python
x
```

```python
y
```

```python
plt.title("Diagram")
plt.scatter(x, y, marker='o')

for i, j in zip(x, y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')

plt.show()
```

```python
reg.fit(x,y)
```

```python
a1 = reg.coef_

a0 = reg.intercept_
```

```python
print('a1 = ' , a1)
print('a0 = ' , a0)
```

```python
#Regression Equation is

print('y =', a0 ,'+', a1,'* x')
```

```python
plt.title("With Regression Equation")
          
y_pred = a0 + (a1*x)

plt.scatter(x,y)
plt.plot(x,y_pred , color ="red")

for i, j in zip(x, y):
 plt.annotate('(%s, %s)' % (i, j), xy=(i, j), textcoords='offset points', xytext=(0,10), ha='center')
plt.show()
```

```python
#Prediction

print('Prediction For x=7' , reg.predict([[7]]))
print('Prediction For x=12' , reg.predict([[12]]))
```
