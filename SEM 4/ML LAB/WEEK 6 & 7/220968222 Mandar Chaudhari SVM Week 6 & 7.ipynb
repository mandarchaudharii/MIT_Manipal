# 5. Split data into separate training and test set 

```python
from sklearn.model_selection import train_test_split
X = df.drop('target', axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
X_test
```

# 6. Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler() 
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

# 7. Run SVM with default hyperparameters 
    Run SVM with rbf kernel and C=1.0 and gamma=auto 
    Run SVM with rbf kernel and C=100.0
    Run SVM with rbf kernel and C=1000.0

```python
from sklearn.svm import SVC
from sklearn import svm
clf1 = svm.SVC()
clf2=svm.SVC(kernel='rbf',C=1,gamma='auto')
clf3=svm.SVC(kernel='rbf',C=100)
clf4=svm.SVC(kernel='rbf',C=1000)
```

# 8. Run SVM with linear kernel
    Run SVM with linear kernel and C=1.0
    Run SVM with linear kernel and C=100.0
    Run SVM with linear kernel and C=1000.0

```python
clf5=svm.SVC(kernel='linear')
clf6=svm.SVC(kernel='linear',C=1)
clf7=svm.SVC(kernel='linear',C=100)
clf8=svm.SVC(kernel='linear',C=1000)
```

# 9. Run SVM with polynomial kernel
    Run SVM with polynomial kernel and C=1.0
    Run SVM with polynomial kernel and C=100.0
    Run SVM with polynomial kernel and C=1000.0

```python
clf9=svm.SVC(kernel='poly')
clf10=svm.SVC(kernel='poly',C=1)
clf11=svm.SVC(kernel='poly',C=100)
clf12=svm.SVC(kernel='poly',C=1000)
```

# 10. Run SVM with sigmoid kernel 
    Run SVM with sigmoid kernel and C=1.0
    Run SVM with sigmoid kernel and C=100.0
    Run SVM with sigmoid kernel and C=1000.0

```python
clf13 = svm.SVC(kernel='sigmoid')
clf14=svm.SVC(kernel='sigmoid',C=1)
clf15=svm.SVC(kernel='sigmoid',C=100)
clf16=svm.SVC(kernel='sigmoid',C=1000)
```

# 11.Compare the train-set and test-set accuracy

# 12.Check for overfitting and underfitting

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score, accuracy_score
classifiers = [svm.SVC(),svm.SVC(kernel='rbf',C=1,gamma='auto'),svm.SVC(kernel='rbf',C=100),svm.SVC(kernel='rbf',C=1000),svm.SVC(kernel='linear'),svm.SVC(kernel='linear',C=1),svm.SVC(kernel='linear',C=100),svm.SVC(kernel='linear',C=1000),svm.SVC(kernel='poly'),svm.SVC(kernel='poly',C=1),svm.SVC(kernel='poly',C=100),svm.SVC(kernel='poly',C=1000),svm.SVC(kernel='sigmoid'),svm.SVC(kernel='sigmoid',C=1),svm.SVC(kernel='sigmoid',C=100),svm.SVC(kernel='sigmoid',C=1000)]
evaluation_results = []
j=1
for i, clf in enumerate(classifiers):
    clf.fit(X_train, y_train)
    y_pred=clf.predict(X_test)
    y_score=clf.decision_function(X_test)
    cm=confusion_matrix(y_test, y_pred)
    cr=classification_report(y_test, y_pred)
    accuracy=accuracy_score(y_test, y_pred)
    error=1-accuracy
    roc_auc=roc_auc_score(y_test, y_score)
    fpr,tpr,_=roc_curve(y_test, y_score)
    evaluation_results.append({"model": f"Model {i+1}","confusion_matrix": cm,"classification_report": cr,"train_accuracy": clf.score(X_train, y_train),"test_accuracy": clf.score(X_test, y_test),"classification_accuracy": accuracy,"classification_error": error,"roc_auc": roc_auc,"roc_curve": (fpr, tpr)})
    print("Model",j,"completed")
    j+=1
```

```python
i=1
for result in evaluation_results:
    print("Model",i)
    print(f"Train Accuracy: {result['train_accuracy']:.2f}")
    print(f"Test Accuracy : {result['test_accuracy']:.2f}")
    i+=1
```

# 13. visualize confusion matrix with seaborn heatmap

```python
i=1
for result in evaluation_results:
    print("Model",i)
    plt.figure()
    sns.heatmap(result['confusion_matrix'],annot=True)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.show()
    i+=1
```

# 14. Generate classification report 

```python
i=1
for result in evaluation_results:
    print("Model",i,"Classification Report")
    print(result['classification_report'])
    i+=1
```

# 15. Evaluate classification accuracy

# 16. Evaluate Classification error

```python
i=1
for result in evaluation_results:
    print("Model",i)
    print(f"Classification Accuracy: {result['classification_accuracy']:.2f}")
    print(f"Classification Error: {result['classification_error']:.2f}")
    i+=1
```

# 17. Plot ROC Curve

```python
i=1
for result in evaluation_results:
    print("Model",i)
    plt.figure()
    plt.plot(result['roc_curve'][0], result['roc_curve'][1])
    plt.plot()
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()
    i+=1
```

# 18. Compute ROC AUC

```python
i=1
for result in evaluation_results:
    print("Model",i)
    print(result['roc_auc'])
    i+=1
```

# 19. Stratified k-fold cross-validation with shuffle split with linear kernel

```python
from sklearn.model_selection import StratifiedKFold, ShuffleSplit, cross_val_score
stratified_kfold = StratifiedKFold(n_splits=3, shuffle=True)
shuffle_split = ShuffleSplit(n_splits=3, test_size=0.2, random_state=42)
cv_scores = cross_val_score(clf5, X, y, cv=stratified_kfold, n_jobs=-1)
print("Cross-validation scores: {}".format(cv_scores))
print("Mean accuracy: {:.2f}".format(np.mean(cv_scores)))
```

# 21. Stratified k-fold cross-validation with shuffle split with rbf kernel 

```python
stratified_kfold = StratifiedKFold(n_splits=3, shuffle=True)
shuffle_split = ShuffleSplit(n_splits=3, test_size=0.2, random_state=42)
cv_scores = cross_val_score(clf2, X, y, cv=stratified_kfold, n_jobs=-1)
print("Cross-validation scores: {}".format(cv_scores))
print("Mean accuracy: {:.2f}".format(np.mean(cv_scores)))
```

# 22. Stratified k-fold cross-validation with shuffle split with polynomial kernel

```python
stratified_kfold = StratifiedKFold(n_splits=3, shuffle=True)
shuffle_split = ShuffleSplit(n_splits=3, test_size=0.2, random_state=42)
cv_scores = cross_val_score(clf9, X, y, cv=stratified_kfold, n_jobs=-1)
print("Cross-validation scores: {}".format(cv_scores))
print("Mean accuracy: {:.2f}".format(np.mean(cv_scores)))
```
